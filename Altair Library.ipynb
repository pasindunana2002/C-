```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import altair as alt
```

```python
from sklearn.datasets import load_wine
```

```python
from sklearn.datasets import load_iris
```

```python
df = pd.read_csv("loan.csv")
```

```python
df.head(4)
```

```python
alt.Chart(df).mark_point().encode(
    x='ApplicantIncome',
    y='LoanAmount',
    color='Loan_Status')
```

```python
alt.Chart(df).mark_point(
  size = 20,
  color = "tomato"

).encode(
    x=alt.X('ApplicantIncome', title = "Income"),
    y=alt.Y('LoanAmount', title = "Loan")
).properties(
    title = "Income vs Loan",
    width=500,
    height=500
).interactive()
```

```python
alt.Chart(df).mark_bar(
  size = 20,
  color = "blue"

).encode(
    x=alt.X('ApplicantIncome', title = "Income"),
    y=alt.Y('LoanAmount', title = "Loan")
).properties(
    title = "Income vs Loan",
    width=500,
    height=500
).interactive()
```
